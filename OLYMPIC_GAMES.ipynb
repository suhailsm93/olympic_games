```python
import pandas as pd
import numpy as np
df = pd.read_csv("summer.csv")
```

```python
#question1.In how many cities Summer Olympics is head so far?

print(df["City"].unique())
print("Number of cities participated in olympics : ",len(df["City"].unique()))
```

```python
#question2.Which sport is having most number of Gold Medals so far?(top5)

goldmedals =df[(df.Medal == 'Gold')]
goldmedals.groupby('Sport').count()['Medal'].sort_values(ascending = False).head(6)
df[df["Medal"]=="Gold"]["Sport"].value_counts().head().plot.bar()
```

```python
#question3.Which sport is having most number of medals so far?

data = []

for sport in df['Sport'].unique():
    data.append([sport , len(df[df['Sport']  == sport])])

medal=pd.DataFrame(data,columns = ['Sport','freq']).sort_values(by='freq', ascending=False).head()
print(medal)
medal.plot(x = 'Sport', y = 'freq', kind = 'bar', figsize = (5,5))
```

```python
#question4.Which player has won most number of medals?( Top 5 )

df["Athlete"].value_counts().head().plot.bar()
```

```python
#question5,Which player has won most number Gold Medals of medals?
gold = df[df['Medal'] == 'Gold']

data = []

for athlete in gold['Athlete'].unique():
    data.append([athlete , len(gold[gold['Athlete']  == athlete])])

goldMedal=pd.DataFrame(data,columns = ['Athlete','freq']).sort_values(by='freq', ascending=False).head()
print(goldMedal)
goldMedal.plot(x = 'Athlete', y = 'freq', kind = 'bar', figsize = (5,5))
```

```python
#question6.In which year India won first Gold Medal in Summer Olympics?

ind = df[(df.Country == 'IND') & (df.Medal == 'Gold')]
y = ind['Year'].head(1)
y
```

```python
#question7.Which event is most popular in terms on number of players?

df.groupby('Event').count()['Athlete'].sort_values(ascending = False).head().plot.bar(figsize = (5,5))
```

```python
#question8.Which sport is having most Female Gold Medalists? ( Top 5 )

df[(df["Gender"]=="Women") & (df["Medal"]=="Gold")]["Sport"].value_counts().head().plot.bar()
```
